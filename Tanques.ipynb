Al segundo tanque, que inicialmente esta lleno hasta tres cuartos de agua pura, tiene agua pura inicialmente, le llega salmuera del primer tanque y le sale a razón de 3 galones por minuto hacia el tercer tanque la mezcla.

**2**. Determinar las ecuaciones para cada tanque.

Donde C_ini:  Concentración inicial

Donde f=flow1=flow2=flow3: Flujo

Donde 

*   V00= Volúmen tanque A
*   V01= Volúmen tanque B
*   V02= Volúmen tanque C



Analisis

Datos



	
  ✔ V00= 200 gal  --------  ✔A(0)=70 lib/gal

  ✔ V01= 200 gal   --------  	 	 		✔B(0)=0 lib/gal

  ✔V02= 200 gal  		  --------   	  ✔C(0)=50 lib/gal


```
def model(y,t):
  dydt=(-2*y[0]/(100+t), 2*y[0]/(100+t)-3*y[1]/(150-t), 3*y[1]/(150-t)-3*y[2]/(140))
  return dydt
y0=[70,0,50]

t=np.linspace(0,100,num=80)
y=odeint(model,y0,t)
plt.xlabel("Tiempo en minutos")
plt.ylabel("Libras de Sal")
plt.plot(t,y[:,0])
plt.plot(t,y[:,1])
plt.plot(t,y[:,2])
plt.scatter(t,y[:,0],label="Tanque A")
plt.scatter(t,y[:,1],label="Tanque B")
plt.scatter(t,y[:,2],label="Tanque C")
plt.legend()
y

```
